```python
import sys
import os
import this
sys.path.append(os.path.abspath("../src"))
from functions import analizar_texto


zen = this.s

lineas_limpias = zen.splitlines()[2:]
texto_para_analizar = "\n".join(lineas_limpias)


lineas, palabras, prom, filtradas = analizar_texto(texto_para_analizar)


print(f"Total de líneas: {lineas}")
print(f"Total de palabras: {palabras}")
print(f"Promedio de palabras por línea: {prom:.2f}")
print(f"\nLíneas por encima del promedio ({prom:.2f} palabras):")
for linea, cant in filtradas:
    print(f"- \"{linea}\" ({cant} palabras)")
```
